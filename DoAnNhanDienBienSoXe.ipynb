```python
from ultralytics import YOLO
import cv2
from PIL import Image
import numpy as np

# Load the model
model = YOLO("/content/runs/detect/train/weights/best.pt")

# Open the video file
video_path = '/content/drive/MyDrive/21DTH2B_DoAnNhanDienBienSoXe_TaQuangHai_2100008852/dataset/data_test/video_test.mp4'
cap = cv2.VideoCapture(video_path)

# Get video properties
frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = int(cap.get(cv2.CAP_PROP_FPS))

# Define codec and create VideoWriter object to save output video
output_path = '/content/output_video.mp4'
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter(output_path, fourcc, fps, (frame_width, frame_height))

while True:
    ret, frame = cap.read()

    # Break the loop if video ends
    if not ret:
        print("End of video reached or error reading frame.")
        break

    # Run YOLO detection
    results = model(frame)

    # Plot and save each frame
    imarray = results[0].plot()  # Using results[0] to access the first result (assuming one frame)
    out_frame = cv2.cvtColor(imarray, cv2.COLOR_RGB2BGR)  # Convert back to BGR for OpenCV
    out.write(out_frame)  # Write the frame to the output video

# Release video objects
cap.release()
out.release()
cv2.destroyAllWindows()

print("Video processing complete!")
```
